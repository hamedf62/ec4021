```python
import requests

def get_prayettimes(cityCode):
    result = requests.get(f"https://prayer.aviny.com/api/prayertimes/{cityCode}")
    return result.json()

get_prayettimes(14)
```

```python
london = get_prayettimes(125)
london
```
